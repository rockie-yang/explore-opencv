# Color Space

There are different way to encode color which called color space. 

The following color spaces are command used

- RGB
- YUV
- HSV

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread("images/input.jpg")
rows, cols, channels = img.shape
```

## Color Space Conversion

OpenCV support color space conversion with function `cv2.cvtColor`

Here we list all the possible conversions

```python
[x for x in dir(cv2) if x.startswith("COLOR_")]
```

## Convert  BGR to YUV

Convert BGR to YUV.  BGR is the default color space by OpenCV, but not RGB.

After convert to YUV, It feels like distorted

```python
yuv_img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
plt.imshow(yuv_img)
```

## Split and Merge Channels

We can split yuv channel and show it separately

we can also merge back with `cv2.merge`

```python
y,u,v = cv2.split(yuv_img)
```

```python
plt.imshow(y, cmap="gray")
```

```python
plt.imshow(u, cmap="gray")
```

```python
plt.imshow(v, cmap="gray")
```

## Split to GBR

```python
b, g, r = cv2.split(img)
zeros = np.zeros([rows, cols], dtype=np.uint8)
```

```python
# Matplot expet RGB format, so we merge as RGB channels
only_green = cv2.merge((zeros, g, zeros))
plt.imshow(only_green)
```

```python
only_blue = cv2.merge((zeros, zeros, b))
plt.imshow(only_blue)
```

```python
only_red = cv2.merge((r, zeros, zeros))
plt.imshow(only_red)
```
