```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.cvtColor(cv2.imread("images/input.jpg"), cv2.COLOR_BGR2RGB)
num_rows, num_cols, num_channels = img.shape
```

```python
# center point: (num_cols/2, num_rows/2)
# 30 degrees
# shrink to 0.7 of orginal size
rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), 30, 0.7)
print(rotation_matrix)
img_rotation = cv2.warpAffine(img, rotation_matrix, (num_cols, num_rows))
plt.imshow(img_rotation)
```
