```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.cvtColor(cv2.imread("images/sign_input.png"), cv2.COLOR_BGR2RGB)
rows, cols, channels = img.shape
plt.imshow(img)
```

## Apply Identity Transformation

The output image is exactly as input image

```python
identity = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])
output = cv2.filter2D(img, -1, identity)
plt.imshow(output)
```

## Apply 3 x 3 Filter to blur the image

```python
one3x3 = np.ones((3, 3), np.float32)
output = cv2.filter2D(img, -1, one3x3/9)
plt.imshow(output)
```

## Apply 5 x 5 kernel to Blur Image

```python
one5x5 = np.ones((5, 5), np.float32)
output = cv2.filter2D(img, -1, one5x5/25)
plt.imshow(output)
```

## Use blur funtion to blur image

```python
output = cv2.blur(img, (7, 7))
plt.imshow(output)
```
