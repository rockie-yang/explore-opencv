## Frequency in Images

Frequency in images are the slope of change.

- Plain image has lower frequency
- Sharp change has high requency

Blur can be treated as low frequency filter

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.cvtColor(cv2.imread("images/tree_input.png"), cv2.COLOR_BGR2RGB)
rows, cols, channels = img.shape
plt.imshow(img)
```

```python
def blur(size=15):
    kernel_motion_blur = np.zeros((size, size))
    kernel_motion_blur[int((size-1)/2), :] = np.ones(size)
    kernel_motion_blur /= size
    print(kernel_motion_blur.shape)

    print(kernel_motion_blur)
    output = cv2.filter2D(img, -1, kernel_motion_blur)
    plt.imshow(output)
```

```python
blur(3)
```

```python
blur(7)
```
