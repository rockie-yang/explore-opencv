```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.cvtColor(cv2.imread("images/input.jpg"), cv2.COLOR_BGR2RGB)
num_rows, num_cols, num_channels = img.shape
```

## Resize Image

- Use INTER_LINEAR or INTER_CUBIC when enlarge images
- INTER_CUBIC is more computational expensive then INTER_LINEAR, while have better quality
- Use INTER_AREA when shrink images

```python
img_scaled = cv2.resize(img, dsize=None, fx=1.2, fy=1.2, interpolation=cv2.INTER_LINEAR)
print("scaled from ", img.shape, img_scaled.shape)
plt.imshow(img_scaled)
```

```python
img_scaled = cv2.resize(img, dsize=None, fx=1.2, fy=1.2, interpolation=cv2.INTER_CUBIC)
print("scaled from ", img.shape, img_scaled.shape)
plt.imshow(img_scaled)
```

```python
img_scaled = cv2.resize(img, dsize=(500, 500), interpolation=cv2.INTER_LINEAR)
print("scaled from ", img.shape, img_scaled.shape)
plt.imshow(img_scaled)
```
