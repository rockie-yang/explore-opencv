```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread("images/input.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
num_rows, num_cols, num_channels = img.shape
```

```python
src_points = np.float32([[0, 0], [num_cols - 1, 0], [0, num_rows - 1]])
dst_points = np.float32([[0, 0], [int(0.6*(num_cols-1)), 0], [int(0.4 * (num_cols - 1)), num_rows-1]])
print(src_points)
print(dst_points)
affine_matrix = cv2.getAffineTransform(src_points, dst_points)
print(affine_matrix)

img_output = cv2.warpAffine(img, affine_matrix, (num_cols, num_rows))
plt.imshow(cv2.cvtColor(img_output, cv2.COLOR_BGR2RGB))
```
