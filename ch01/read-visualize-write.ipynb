```python
import cv2
from matplotlib import pyplot as plt
```

## Read Images

OpenCV can read and decode image for varies formats. 

After read it will be plain numpy array

```python
img = cv2.imread("images/input.jpg")
print(type(img))
print(img.shape)
```

## Visualize Images

OpenCV can visusalize image it self. 

while it does not work out of box nicely in Jupyter

```python

cv2.imshow("input image", img)

# This will show the image, while it will be in wait key status, and has to be forced killed
# cv2.waitKey()
```

## Visualize with Matplotlib

the image is an 3 channel image, while opencv encoded as BGR color

In order to show it with matplotlib, we shall converte it to RGB

```python
plt.imshow(img)
```

```python
# We can use this trick to convert to RGB
img2 = img[:,:,::-1]
plt.imshow(img2)
```

```python
# Or use cvtColor to change from BGR2RGB
RGB_im = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_im)
```

## Visualize it as gray image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray_img.shape)
plt.imshow(gray_img, cmap="gray")
```

## Save Images

```python
# Save the gray image
cv2.imwrite("images/output_gray.jpg", gray_img)
```

```python
# Save the image as png
cv2.imwrite("images/output_gray.png", gray_img, [cv2.IMWRITE_PNG_COMPRESSION])
```

```python
# opencv can detect the extension as well
cv2.imwrite("images/output.png", img)
```

## Read back Images

```python
# by default, opencv will read as cv2.IMREAD_COLOR mode
# which has 3 channels
# each channel will have same value if it's a gray image
input_gray=cv2.imread("images/output_gray.jpg")
print(input_gray.shape)
print(input_gray[1, 1, 0], input_gray[1, 1, 1], input_gray[1, 1, 2])
plt.imshow(input_gray)
```

```python
input_gray=cv2.imread("images/output_gray.jpg", cv2.IMREAD_GRAYSCALE)
print(input_gray.shape)
print(input_gray[1, 1])
# when the data is gray we need specify the color map as gray
# otherwise, it will interpret differently
plt.imshow(input_gray, cmap="gray")
```
