```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.cvtColor(cv2.imread("images/input.jpg"), cv2.COLOR_BGR2RGB)
rows, cols, channels = img.shape
```

## Shift Images

- x shift 70
- y shift 110

The other two values [1, 0] and [0, 1] are fixed to annotate which axis

The size keep the same in this case

```python
translation_matrix = np.float32([ [1,0,70], [0,1,110] ])
img_translation = cv2.warpAffine(img, translation_matrix, (cols, rows), cv2.INTER_LINEAR)
plt.imshow(img_translation)
```

## Shift Image but avoid crops



```python
translation_matrix = np.float32([ [1,0,70], [0,1,110] ])
img_translation = cv2.warpAffine(img, translation_matrix, (cols + 70, rows + 110), cv2.INTER_LINEAR)
plt.imshow(img_translation)
```

## Warp Affine

\begin{equation*}
T = \begin{vmatrix}
1 & 0 & t_x \\
0 & 1 & t_y
\end{vmatrix}
\end{equation*}

```python
translation_matrix = np.float32([ [1,0,70], [0,1,0] ])
img_translation = cv2.warpAffine(img, translation_matrix, (cols, rows))
plt.imshow(img_translation)
```
